```python
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/shootings.csv"
data = pd.read_csv(url)
```

```python
print(" First 5 Rows of the Dataset:")
display(data.head())
```

```python
print("\nDataset Shape:", data.shape)
```

```python
print("\n Missing Values in Each Column:")
data['manner_of_death'].fillna(data['manner_of_death'].mode(),inplace=True)
armsna=data['arms_category'].mode()
print(armsna)
data['arms_category'].fillna(armsna,inplace=True)
data['race'].fillna(data['race'].mode(),inplace=True)

print(data.isnull().sum())
```

```python
print(" Summary Statistics:")
display(data.describe(include='all'))
```

```python
# 1. Finding and Listing Non-Numeric Values
print("="*50)
print("Step 1: Non-Numeric Values Analysis")
print("="*50)

# Identify non-numeric columns
non_numeric_count = data.select_dtypes(include=['object']).nunique()
print("\nNumber of unique non-numeric values in each column:")
print(non_numeric_count)

# Show unique values in categorical columns
non_numeric_values = data.select_dtypes(include=['object']).apply(lambda col: col.unique())
print("\nNon-numeric values in each column:")
print(non_numeric_values)
```

```python
data['age'].fillna(data['age'].median(), inplace=True)
```

```python
bins = [0, 20, 40, 60, 80]
labels = ['0-20', '21-40', '41-60', '61-80']
data['age_binned'] = pd.cut(data['age'], bins=bins, labels=labels)

# Display the updated dataframe with binned categories
print("Dataset with 'age' discretized into bins:")
print(data[['age', 'age_binned']].head())
```

```python
Q1 = data['age'].quantile(0.25)
Q3 = data['age'].quantile(0.75)
IQR = Q3 - Q1

# Defining outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detecting outliers
outliers = data[(data['age'] < lower_bound) | (data['age'] > upper_bound)]
print(f"Number of outliers detected in 'duration': {len(outliers)}")

# Optionally remove outliers (uncomment the line below to remove them)
# data = data[(data['duration'] >= lower_bound) & (data['duration'] <= upper_bound)]
print(f"Dataset shape after outlier handling: {data.shape}")
```

## Plots for Univariate,Bivariate and Multivariate

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")


plt.figure(figsize=(12, 5),dpi=1000)
sns.boxplot(data=data, x="arms_category", y="age", palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Age Distribution by Arms Category")
plt.xlabel("Arms Category")
plt.ylabel("Age")
plt.show()


plt.figure(figsize=(10, 5))
sns.countplot(data=data, x="race", hue="threat_level", palette="viridis")
plt.title("Threat Level Distribution by Race")
plt.xlabel("Race")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(10, 6))
corr_matrix = data.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python


# 8. Pair Plot - Age, Mental Illness, Flee Status, Threat Level
plt.figure(figsize=(10,10),dpi=1000)
shootings_subset = data[['age', 'signs_of_mental_illness', 'flee', 'threat_level']]
sns.pairplot(shootings_subset, hue='flee', palette='coolwarm')
plt.suptitle("Pair Plot of Key Features - 2124", y=1.02)
plt.show()
```
